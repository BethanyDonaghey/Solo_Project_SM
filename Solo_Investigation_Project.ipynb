```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.05
eta = 0.04
beta = 0.0002
gamma = 0.0002

M0 = 500
F0 = 600

t0 = 0
t_end = 10
dt = 0.1

t_values = np.arange(t0, t_end + dt, dt)

M = np.zeros(len(t_values))
F = np.zeros(len(t_values))

M[0] = M0
F[0] = F0

for i in range(len(t_values) - 1):
    dMdt = -alpha * M[i] + beta * M[i] * F[i]
    dFdt = -eta * F[i] + gamma * M[i] * F[i]

    M[i + 1] = M[i] + dt * dMdt
    F[i + 1] = F[i] + dt * dFdt

plt.figure()
plt.plot(t_values, M, label="Male population M(t)")
plt.plot(t_values, F, label="Female population F(t)")
plt.xlabel("Time (years)")
plt.ylabel("Population Size")
plt.title("Blue Whale Population Model")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


alpha = 0.05
eta = 0.04
beta = 0.00005
gamma = 0.00005

K = 5000

M0 = 500
F0 = 600

t0 = 0
t_end = 70
dt = 0.1

t_values = np.arange(t0, t_end + dt, dt)

M = np.zeros(len(t_values))
F = np.zeros(len(t_values))

M[0] = M0
F[0] = F0

for i in range(len(t_values) - 1):
    total = M[i] + F[i]
    limiting_factor = 1 - total / K

    dMdt = -alpha * M[i] + beta * M[i] * F[i] * limiting_factor
    dFdt = -eta * F[i] + gamma * M[i] * F[i] * limiting_factor

    M[i + 1] = M[i] + dt * dMdt
    F[i + 1] = F[i] + dt * dFdt

plt.figure()
plt.plot(t_values, M, label="Male population M(t)")
plt.plot(t_values, F, label="Female population F(t)")
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.title("Blue Whale Population Model with Resource Limitation (K = 5000)")
plt.legend()
plt.show()
```
